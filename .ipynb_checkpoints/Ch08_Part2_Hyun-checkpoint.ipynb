### 8.4.3 변이형 오토인코더(VAE)

---

* 오토인코더는 입력을 저차원 잠재 공간으로 인코딩한후 디코딩하여 복원하는 네트워크
* VAE는 딥러닝과 베이지안 추론의 아이디어를 혼합한 오토인코더

* #### 고전적인 오토인코더
    1. 이미지를 입력받아 인코더 모듈을 사용하여 잠재 벡터 공간으로 매핑
    2. 디코더 모듈을 사용 원본 이미지와 동일한 차원으로 복원하여 출력
    
    ![image.png](.\Fig\fig_08_12.png)

* 오토인코더는 입력 이미지와 동일한 이미지를 타깃 데이터로 사용하여 훈련(input img = label data)
* 코딩(인코더의 출력)에 여러 제약을 가하면 오토인코더가 더, 덜 흥미로운 잠재 공간의 표현을 학습.
* you can get the autoencoder to learn more-or-less interesting latent representations of the data.
* 일반적인 제약 방법 : 코딩이 저차원, 희소하도록 만듦
* <b> 하지만 유용하거나, 좋은 잠재 공간을 생성하기 어려움</b>

* #### VAE는 통계 기법을 추가하여 연속적이고 구조적인 잠재 공간을 학습함
* 입력 이미지를 잠재 공간의 고정된 코딩으로 압축하는 대신 VAE는 이미지를 어떤 통계 분포의 파라미터로 변환. (분포의 parameter : mean & variance)
* 통계 분포의 파라미터로 변환이 의미 하는것 : 인코딩과 디코딩하는 동안 무작위성이 필요하다는것
* VAE는 평균과 분산 파라미터를 사용하여 이 분포에서 무작위로 하나의 샘플을 추출후 디코딩하여 원본입력으로 복원
* 이러한 무작위한 과정은 안정성 향상, 잠재공간 어디서든 의미있는 표현을 인코딩 하도록 함

![title](.\fig\Fig_08_13_kr.png)

#### 기술적 VAE 작동원리

1. 인코더 모듈이 입력 샘플 input_img를 잠재 공간의 두 파라미터 z_mean과 z_log_var로 변환
2. 입력 이미지가 생성되었다고 가정한 잠재 공간의 정규 분포에서 포인트 z를 z = z_mean+exp(0.5 * z_log_var) * epsilon 처럼 무작위로 샘플링
3. 디코더 모듈은 잠재 공간의 이 포인트를 원본 입력 이미지로 매핑하여 복원

* 작고 무작위적인 epsilon값 때문에 input_img를 인코딩한 잠재 공간의 위치(z_mean)에 가까운 포인트는 input_img와 비슷한 이미지로 디코딩됨 -> 잠재 공간에서 가까운 2개의 포인트는 아주 비슷한 이미지로 디코딩됨 -> 이러한 연속성은 잠재공간에서 방향이 의미있도록 해줌 -> 잠재공간이 구조적이고 벡터로 다루기에 적합해짐

VAE의 파라미터는 2개의 loss function으로 훈련
1. 디코딩된 샘플이 원본 입력과 동일하도록 만드는 재구성 손실(reconstruction loss)
2. latent spaces를 잘 형성하고 훈련데이터에 overfitting 을 줄이는 규제손실(regularization loss)

#### 간략한 케라스의 VAE 구현

입력을 mean. variance parameter로 인코딩<br>
z_mean, z_log_var = encoder(input_img)<br>
무작위의 작은 epsilon 값을 사용하여 잠재공간의 포인트를 추출<br>
z = z_mean + exp(0.5 * z_log_var)* epsilon<br>
z를 이미지로 디코딩<br>
reconstructed_img = decoder(z)<br>
입력이미지와 재구성 이미지를 매핑한 오토인코더 모델을 만듦<br>
model = Model(input_img, reconstructed_img)<br>

이 모델을 reconstruction loss, regularization loss를 사용하여 훈련.

### 인코더 네트워크 구현 

```python
# 이미지를 잠재 공간상 환륜 분포 파라미터로 매핑하는 인코더 네트워크

import keras
from keras import layers
from keras import backend as K
from keras.models import Model
import numpy as np
```

```python
img_shape = (28, 28, 1)
batch_size = 16
latent_dim = 2  # 잠재 공간의 차원: 2D 평면

input_img = keras.Input(shape=img_shape)

x = layers.Conv2D(32, 3,
                  padding='same', activation='relu')(input_img)
x = layers.Conv2D(64, 3,
                  padding='same', activation='relu',
                  strides=(2, 2))(x)
x = layers.Conv2D(64, 3,
                  padding='same', activation='relu')(x)
x = layers.Conv2D(64, 3,
                  padding='same', activation='relu')(x)
shape_before_flattening = K.int_shape(x)

x = layers.Flatten()(x)
x = layers.Dense(32, activation='relu')(x)
# encoder.summary()
z_mean = layers.Dense(latent_dim)(x)
z_log_var = layers.Dense(latent_dim)(x)
```

```python
encoder.summary()
```

* z_mean과 z_log_var를 사용하는 코드
* 두 파라미터가 input_img를 생성한 통계분포의 파라미터라 가정, 잠재공간 포인트 z를 생성 
* z_mean, z_log_var를 Lambda층으로 감쌈

```python
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim),
                              mean=0., stddev=1.)
    return z_mean + K.exp(z_log_var) * epsilon

z = layers.Lambda(sampling)([z_mean, z_log_var])
```

### 디코더 구현

```python
decoder_input = layers.Input(K.int_shape(z)[1:])

# 입력을 업샘플링
x = layers.Dense(np.prod(shape_before_flattening[1:]),activation='relu')(decoder_input)
x = layers.Reshape(shape_before_flattening[1:])(x)
x = layers.Conv2DTranspose(32,3,padding='same',activation='relu',strides=(2,2))(x)
x = layers.Conv2D(1,3,padding='same',activation='sigmoid')(x)

decoder = Model(decoder_input,x)

z_decoded = decoder(z)
```

```python
decoder.summary()
```

일반적인 샘플 기준의 함수인 loss(y_true,y_pred)형태는 VAE의 이중손실(reconstruction loss과 regularization loss)에 맞지 않음

```python
# VAE loss 를 계산하기 위한 layer
class CustomVariationalLayer(keras.layers.Layer):

    def vae_loss(self, x, z_decoded):
        x = K.flatten(x)
        z_decoded = K.flatten(z_decoded)
        xent_loss = keras.metrics.binary_crossentropy(x, z_decoded)
        kl_loss = -5e-4 * K.mean(
            1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
        return K.mean(xent_loss + kl_loss)

    def call(self, inputs):
        x = inputs[0]
        z_decoded = inputs[1]
        loss = self.vae_loss(x, z_decoded)
        self.add_loss(loss, inputs=inputs)
        # 출력 값을 사용하지 않습니다
        return x

# 입력과 디코딩된 출력으로 이 층을 호출하여 모델의 최종 출력을 얻습니다
y = CustomVariationalLayer()([input_img, z_decoded])
```

loss를 다루는 층을 직접 다루기 때문에 compile 메서드에서 손실을 지정하지 않음

```python
from keras.datasets import mnist
```

```python
vae = Model(input_img,y)
```

```python
vae.compile(optimizer='rmsprop',loss=None)
vae.summary()

(x_train,_),(x_test,y_test)= mnist.load_data()

x_train = x_train.astype('float32')/255.
x_train = x_train.reshape(x_train.shape + (1,))
x_test = x_test.astype('float32')/255.
x_test = x_test.reshape(x_test.shape+(1,))

vae.fit(x = x_train,y=None,
        shuffle=True,
        epochs=10,
        batch_size = batch_size,
        validation_data=(x_test,None))
```

### 8.4.4 정리
* 딥러닝으로 통계정보를 담은 latent space를 학습한후 이미지를 생성할수 있음
* 잠재 공간에서 포인트를 샘플링하고 디코딩하면 이미지를 생성할수 있음 주요 방법은 VAE와 GAN이 있음
* VAE는 매우 구조적이고 연속적인 잠재 공간의 표현을 생성, latent space안에서 일어나는 이미지 변형종류의 이미지 변형 작업에 어울림
    * 다른 얼굴로 바꾸기
    * 찌푸린 얼굴을 웃는 얼굴로 바꾸기 등등
* GAN은 실제 같은 단일 이미지를 생성할수 있지만 구조적이고 연속적인 잠재 공간을 만들지 못함

## 8.5 적대적 생성 신경망(Generative Adversarial Networks, GAN) 소개
---

* 가짜 피카소를 만드는 위조범<br>
* 진짜 피카소그림과 가짜 피카소 그림을 파는 판매상<br>
* 처음엔 형편없이 위조하지만 판매상이 피드백을 해주고 위조범은 점점 능숙하게 위조하게 되고 판매상은 위조품을 구분하는데 전문가가 됨<br>

* 생성자 네트워크 (generator network) : 랜덤벡터를 입력으로 받아 이를 합성된 이미지로 디코딩<br>
* 판별자 네트워크 (discriminator network) : 이미지(실제 또는 합성 이미지)를 입력으로 받아 훈련 세트에서 온 이미지인지, 생성자 네트워크가 만든 이미지인지 판별함.

* generator network : discriminator network를 속이도록 훈련, 훈련이 계속될수록 점점 더 실제와 같은 이미지 생성
* discriminator network : 생성된 이미지가 실제인지 판별하는 기준을 설정하면서 생성자의 능력 향상에 적응해 갑니다.

* 훈련이 끝난후 generator는 입력 공간에 있는 어떤 포인트를 그럴듯한 이미지로 변환 하지만 <b>이 공간은 연속적이지 않음</b>

![title](.\fig\Fig_08_15.png)

* GAN은 gradient descent를 하며 내려오는 매 단계가 조금씩 공간을 바꿈, GAN을 만들기 위해선 모델 구조와 훈련 파라미터를 많이 조정해주어야함

## 8.5.1 GAN 구현항법
---

* 생성자와 판별자가 심층 컨브넷. 특히 생성자에서 이미지 업샘플링을 위해 Conv2DTranspose layer를 사용
* CIFAR10 데이터셋의 이미지로 GAN 훈련
    * CIFAR10 데이터셋 : 32x32 RGB 이미지 5만개로 이루어져 있고 10개의 클래스를 가짐 (10개의 클래스중 'frog'클래스의 이미지 사용)
  

#### GAN의 구조

1. generator 네트워크는 (latent_dim,)크기의 벡터를 (32,32,3) 크기의 이미지로 매핑
2. discriminator 네트워크는 (32,32,3)크기의 이미지가 진자일 확률을 추정하여 이진 값으로 매핑
3. generator와 discriminator를 연결하는 gan 네트워크를 생성 gan(x) = discriminator(generator(x)) 이 네트워크는 latent space의 벡터를 discriminator의 평가로 매핑
   판별자는 생성자가 잠재 공간의 벡터를 디코딩한것이 얼마나 현실적인지 평가
4. True/False 레이블과 함께 진자 이미지와 가짜 이미지 샘플을 사용하여 판별자를 훈련
5. 생성자를 훈련하려면 gan 모델의 손실에 대한 생성자 weight의 gradient를 사용 == 매 단계마다 생성자에 의해 디코딩괸 이미지를 판별자가 True로 분류하도록 만드는 방향으로 생성자의 Weight를 이동

## 8.5.2 훈련 방법
---
* GAN을 훈련하고 튜닝하는것은 매우 어려움
* 몇가지 기법 존재 : 이론에 바탕하는 지침이 아닌 경험을 통해 발견
* 8장의 GAN 모델을 구현하는데 사용할 몇가지 기법
    1. 생성자의 마지막 활성화로 Sigmoid가 아닌 tanh 함수 사용.
    2. 균등분포(uniform distribution)이 아닌 가우시안 분포(Gaussian distribution)사용
    3. 무작위성을위해 Dropout 또는 레이블에 random noise 추가
    4. use LeakyReLU instead of ReLU, use strided convolution instead of MaxPooling
    5. generator 에서 픽셀 공간을 균일하게 다루지 못할경우 체스판 모양이 종종 나타남 이를 해결하기위해 생성자, 판별자에서 스트라이드 Conv2DTranpose나 Conv2D를 사용할 때 스트라이드 크기로 나누어질 수 있는 커널 크기를 사용

## 8.5.3 생성자 (generator)
---
* 이미지를 생성하는 generator 모델

```python
# 코드 8-29 GAN generator network
import keras
from keras import layers
import numpy as np
```

```python
latent_dim = 32
height = 32
width = 32
channels = 3

generator_input = keras.Input(shape= (latent_dim,))
x = layers.Dense(128*16*16)(generator_input)
x = layers.ReLU()(x)
x = layers.Reshape((16,16,128))(x)

x = layers.Conv2D(256,5,padding='same')(x)
x = layers.ReLU()(x)

x = layers.Conv2DTranspose(256,4,strides=2,padding='same')(x)
x = layers.ReLU()(x)

x = layers.Conv2D(256,5,padding='same')(x)
x = layers.ReLU()(x)
x = layers.Conv2D(256,5,padding='same')(x)
x = layers.ReLU()(x)

x = layers.Conv2D(channels,7,activation='tanh',padding='same')(x)
generator = keras.models.Model(generator_input,x)
generator.summary()
```

## 8.5.4 판별자( discriminator)
---
후보이미지 (True/False)를 입력받은 후 2개의 클래스로 분류하는 Discriminator 모델

```python
discriminator_input = layers.Input(shape=(height,width,channels))

x = layers.Conv2D(128,3)(discriminator_input)
x = layers.LeakyReLU()(x)
x = layers.Conv2D(128,4,strides=2)(x)
x = layers.LeakyReLU()(x)
x = layers.Conv2D(128,4,strides=2)(x)
x = layers.LeakyReLU()(x)
x = layers.Conv2D(128,4,strides=2)(x)
x = layers.LeakyReLU()(x)
x = layers.Flatten()(x)

x = layers.Dropout(0.5)(x)

x = layers.Dense(1,activation='sigmoid')(x)  # classification

discriminator = keras.models.Model(discriminator_input,x) # (32,32,3) 크기의 입력을 이진 분류 결정로 변환하는 판별자 모델

discriminator_optimizer = keras.optimizers.RMSprop(lr=0.0008,
                                                   clipvalue =1.0, # 옵티마이저에서 gradient cliping 사용
                                                   decay=1e-8) # 학습률 감쇠
discriminator.compile(optimizer = discriminator_optimizer,
                     loss ='binary_crossentropy')
```

## 8.5.5 적대적 네트워크 (Generative Adversarial Network)
---
* 마지막으로 generator 과 discriminator을 연결하여 GAN을 설정</br>
* 훈련에 사용되는 타깃레이블은 항상 True img </br>
* gan 훈련의 목표 : generator가 만든 이미지를 보고 discriminator가 진짜라고 예측을 하도록 만드는것</br>
* <b> 판별자가 학습되지 않도록 하는것이 중요</b>


```python
# 코드 8-31 GAN

discriminator.trainable = False

gan_input = keras.Input(shape=(latent_dim,))
gan_output = discriminator(generator(gan_input))
gan = keras.models.Model(gan_input,gan_output)

gan_optimizer = keras.optimizers.RMSprop(lr = 0.0004, clipvalue = 1.0 , decay=1e-8)
gan.compile(optimizer = gan_optimizer, loss='binary_crossentropy')
```

## 8.5.6 DCGAN 훈련 방법
---
* 매 반복마다 다음을 수행
    1. latent space 에서 무작위로 포인트를 뽑음
    2. 이 무작위 포인트를 사용하여 generator에서 이미지를 생성
    3. 생성된 이미지와 진짜 이미지를 섞음
    4. 진짜와 가짜가 섞인 이미지와 이에 대응하는 타깃을 사용하여 Discriminator를 훈련, 타깃은 True or False
    5. latent space 에서 무작위로 새로운 포인트를 뽑음
    6. 이 랜덤 벡터를 사용하여 gan을 훈련 모든 타깃은 True 로 설정, Discriminator가 생성된 이미지를 모두 True로 예측하도록 generator의 weight를 업데이트

```python
# 코드 8-32 GAN훈련 구현하기
import os
from keras.preprocessing import image
```

```python
(x_train,y_train),(_,_) = keras.datasets.cifar10.load_data()
x_train = x_train[y_train.flatten()==6]

x_train = x_train.reshape((x_train.shape[0],)+ (height,width,channels)).astype('float32')/255.

iterations = 10000
batch_size = 20
save_dir = './datasets/gan_images'
if not os.path.exists(save_dir):
    os.makedirs(save_dir,exist_ok=True)
    
start =0
for step in range(iterations):
    # 잠재 공간에서 무작위로 포인트를 샘플링
    random_latent_vectors = np.random.normal(size=(batch_size,latent_dim))
    # 가짜 이미지를 디코딩
    generated_images = generator.predict(random_latent_vectors)
    
    # 진짜 이미지와 연결
    stop = start + batch_size
    real_images = x_train[start:stop]
    combined_images = np.concatenate([generated_images,real_images])
    
    #진짜와 가짜 이미지를 구분하여 레이블을 합침
    labels = np.concatenate([np.ones((batch_size,1)),np.zeros((batch_size,1))])
    # 레이블에 렌덤 노이즈를 추가
    labels += 0.05 * np.random.random(labels.shape)
    # discriminator를 훈련
    discriminator.trainable = True
    d_loss = discriminator.train_on_batch(combined_images,labels)
    # latent space에서 무작위로 포인트를 샘플링
    random_latent_vectors = np.random.normal(size = (batch_size,latent_dim))
    # 모두 True img 라고 레이블을 생성
    misleading_targets = np.zeros((batch_size,1))
    # generator를 훈련(gan 모델에서 discrminator의 가중치는 동결)
    discriminator.trainable = False
    a_loss = gan.train_on_batch(random_latent_vectors,misleading_targets)
    
    start += batch_size
    if start > len(x_train) - batch_size:
        start = 0
    if step % 100 == 0:
        #모델 가중치 저장
        gan.save_weights('gan.h5')
        
        print('판별자 손실 : ',d_loss)
        print('적대적 손실 : ',a_loss)
        
        img = image.array_to_img(generated_images[0]*255.,scale=False)
        img.save(os.path.join(save_dir,'generated_frog'+str(step)+'.png'))
        
        img = image.array_to_img(real_images[0]*255.,scale=False)
        img.save(os.path.join(save_dir,'real_frog'+str(step)+'.png')) 
```

## 8.5.7 정리
---
* GAN은 generator + discriminator 로 구성됩니다.
    * generator : discriminator를 속이도록 훈련
    * discriminator : generator와 진짜 이미지를 구분하도록 훈련
    * generator는 discriminator에게 정보만 얻을뿐 실제 이미지를 훈련되지 않음
* GAN은 훈련하기 어렵습니다 
    * 고정된 손실 공간에서 수행하는 단순한 gradient descent가 아닌 동적과정이기 때문
    * 많은 튜닝이 필요
* GAN은 매우 실제같은 이미지를 만들 수 있습니다.
    * VAE와 달리 학스보딘 잠재 공간이 깔끔하게 연속된 구조를 가지지 않음

# 8.6 요약
---
* 딥러닝을 사용한 애플리케이션을 배웠습니다, 직접 콘텐츠를 생산할수 있습니다.
    * 한번에 하나의 타임스텝씩 시퀀스데이터 생성, 택스트 생성이나 음표하나생성하는법
    * 딥드림의 작동원리, 컨브넷 층 활성화를 최대화 하기위해 입력 공간에 경사 상승법을 적용
    * 스타일 트렌스퍼 적용법, 콘텐츠 이미지와 스타일 이미지가 연결
    * GAN, VAE를 이용한 이미지 생성법

```python
latent_dim = 32
height = 32
width = 32
channels = 3

generator_input = keras.Input(shape=(latent_dim,))

# 입력을 16 × 16 크기의 128개 채널을 가진 특성 맵으로 변환합니다
x = layers.Dense(128 * 16 * 16)(generator_input)
x = layers.LeakyReLU()(x)
x = layers.Reshape((16, 16, 128))(x)

# 합성곱 층을 추가합니다
x = layers.Conv2D(256, 5, padding='same')(x)
x = layers.LeakyReLU()(x)

# 32 × 32 크기로 업샘플링합니다
x = layers.Conv2DTranspose(256, 4, strides=2, padding='same')(x)
x = layers.LeakyReLU()(x)

# 합성곱 층을 더 추가합니다
x = layers.Conv2D(256, 5, padding='same')(x)
x = layers.LeakyReLU()(x)
x = layers.Conv2D(256, 5, padding='same')(x)
x = layers.LeakyReLU()(x)

# 32 × 32 크기의 1개 채널을 가진 특성 맵을 생성합니다
x = layers.Conv2D(channels, 7, activation='tanh', padding='same')(x)
generator = keras.models.Model(generator_input, x)
generator.summary()
```

```python
discriminator_input = layers.Input(shape=(height, width, channels))
x = layers.Conv2D(128, 3)(discriminator_input)
x = layers.LeakyReLU()(x)
x = layers.Conv2D(128, 4, strides=2)(x)
x = layers.LeakyReLU()(x)
x = layers.Conv2D(128, 4, strides=2)(x)
x = layers.LeakyReLU()(x)
x = layers.Conv2D(128, 4, strides=2)(x)
x = layers.LeakyReLU()(x)
x = layers.Flatten()(x)

# 드롭아웃 층을 넣는 것이 아주 중요합니다!
x = layers.Dropout(0.4)(x)

# 분류 층
x = layers.Dense(1, activation='sigmoid')(x)

discriminator = keras.models.Model(discriminator_input, x)
discriminator.summary()

# 옵티마이저에서 (값을 지정하여) 그래디언트 클리핑을 사용합니다
# 안정된 훈련을 위해서 학습률 감쇠를 사용합니다
discriminator_optimizer = keras.optimizers.RMSprop(lr=0.0008, clipvalue=1.0, decay=1e-8)
discriminator.compile(optimizer=discriminator_optimizer, loss='binary_crossentropy')
```

```python
# 판별자의 가중치가 훈련되지 않도록 설정합니다(gan 모델에만 적용됩니다)
discriminator.trainable = False

gan_input = keras.Input(shape=(latent_dim,))
gan_output = discriminator(generator(gan_input))
gan = keras.models.Model(gan_input, gan_output)

gan_optimizer = keras.optimizers.RMSprop(lr=0.0004, clipvalue=1.0, decay=1e-8)
gan.compile(optimizer=gan_optimizer, loss='binary_crossentropy')
```

```python
# CIFAR10 데이터를 로드합니다
(x_train, y_train), (_, _) = keras.datasets.cifar10.load_data()

# 개구리 이미지를 선택합니다(클래스 6)
x_train = x_train[y_train.flatten() == 6]

# 데이터를 정규화합니다
x_train = x_train.reshape(
    (x_train.shape[0],) + (height, width, channels)).astype('float32') / 255.

iterations = 10000
batch_size = 20
save_dir = './datasets/gan_images/'
if not os.path.exists(save_dir):
    os.mkdir(save_dir)

# 훈련 반복 시작
start = 0
for step in range(iterations):
    # 잠재 공간에서 무작위로 포인트를 샘플링합니다
    random_latent_vectors = np.random.normal(size=(batch_size, latent_dim))

    # 가짜 이미지를 디코딩합니다
    generated_images = generator.predict(random_latent_vectors)

    # 진짜 이미지와 연결합니다
    stop = start + batch_size
    real_images = x_train[start: stop]
    combined_images = np.concatenate([generated_images, real_images])

    # 진짜와 가짜 이미지를 구분하여 레이블을 합칩니다
    labels = np.concatenate([np.ones((batch_size, 1)),
                             np.zeros((batch_size, 1))])
    # 레이블에 랜덤 노이즈를 추가합니다. 아주 중요합니다!
    labels += 0.05 * np.random.random(labels.shape)

    # discriminator를 훈련합니다
    discriminator.trainable = True
    d_loss = discriminator.train_on_batch(combined_images, labels)

    
    # 잠재 공간에서 무작위로 포인트를 샘플링합니다
    random_latent_vectors = np.random.normal(size=(batch_size, latent_dim))

    # 모두 “진짜 이미지"라고 레이블을 만듭니다
    misleading_targets = np.zeros((batch_size, 1))

    # generator를 훈련합니다(gan 모델에서 discriminator의 가중치는 동결됩니다)
    discriminator.trainable = False
    a_loss = gan.train_on_batch(random_latent_vectors, misleading_targets)
    
    start += batch_size
    if start > len(x_train) - batch_size:
      start = 0

    # 중간 중간 저장하고 그래프를 그립니다
    if step % 100 == 0:
        # 모델 가중치를 저장합니다
        gan.save_weights('gan.h5')

        # 측정 지표를 출력합니다
        print('스텝 %s에서 판별자 손실: %s' % (step, d_loss))
        print('스텝 %s에서 적대적 손실: %s' % (step, a_loss))

        # 생성된 이미지 하나를 저장합니다
        img = image.array_to_img(generated_images[0] * 255., scale=False)
        img.save(os.path.join(save_dir, 'generated_frog' + str(step) + '.png'))

        # 비교를 위해 진짜 이미지 하나를 저장합니다
        img = image.array_to_img(real_images[0] * 255., scale=False)
        img.save(os.path.join(save_dir, 'real_frog' + str(step) + '.png'))
```
